## What is systems biology?

One thing Jeff gets really excited by right at the start of the lecture is a [video of a neutrophil chasing around a bacterium](https://youtu.be/gc3O2sKIsX4?list=PLUl4u3cNGP63OI3pSKo8Ha_DFBMxm23xO&t=83) in some blood. This is really amazing! How can a single cell make these complex decisions and chase another cell? This is the sort of behaviour that systems biology aims to understand. 

Systems biology uses simple models from nonlinear dynamics and physics to model simple cells, genes, and cell-interactions.

## A gentle introduction to Gene Regulation

- $ X \rightarrow Y$ means (gene) X activates (gene) Y

- $X \dashv Y$ means X represses Y


### Examples:

A. $\text{aTc} \dashv \text{TetR} \dashv \text{gfp}$

- gfp is green flourescant protein - it is very important for experiments! IDK what the other things stand for. 

Qn. What is the characteristic timescale for these sort of activations/ repressions to occur?

Ans. Of order on the cell generation time, i.e. timescale of the cell cycle. (Which varys depending on the type of cell but is roughly an hour for a bacterium.)

B. Negative autoregulation $\text{aTc} \dashv \text{TetR-gfp} \dashv \text{TetR-gfp}$

TetR-gfp is a *fusion protein*. 

Autoregulation can decreate response time. 

From two genes repressing one another, you can make a toggle switch! This is probably the smallest component of computer memory, it is something a little bit like a [latch](https://www.doc.ic.ac.uk/~eedwards/compsys/memory/latches.html) or something. 

With three genes repressing one another in a circular fashion, you get oscillations as the states mutually repress one another. Something like:

- $\text{gfp} \dashv \text{bfp} \dashv \text{rfp} \dashv \text{gfp}$

This is called a repressilator! It is very cool. But if you do this with colonies of e-coli, they get out of phase with one another. Elowitz came up with this - and through this found about the importance of noise in the decision making of cells. (The three proteins are differently coloured but otherwise basically the same.)

Ulon and Barabasi looked for motifs within gene regulatory networks, and found a lot of interesting ones. We will look at the feed-forward loop motif:

$$ A \rightarrow B \rightarrow C \text{ and } A \rightarrow C$$

A conclusion to all of these things: **Life is noisy**. You can do the same experiment twice and get very different answers just due to noise in molecular scale gene expression. 

Another conclusion: **Life is different if you are small**. A case in point, Reynold's number changes a lot with scale (at bacterial scale, $\mathcal{Re} = 10^{-4}$ and water is sticky like treacle). 

## Turing Patterns and bigger things

We will look at these in the course; we are also going to look at systems biology in ecology (we'll look at more papers by Uri Alon, including a paper on the evolution of optimal expression level). 

- ### Stochasticity in systems biology

Qn. Why do we need to consider stochasticity in populations of approx. $10^9$ individuals? Should this not just wash out (like in thermodynamics)?

Ans. Every mutant starts out as a single individual. This means that there is always a time where populations are subject to stochastic fluctuations, which can be very important. 

- ### Fitness Landscapes
This is another important concept - with enough individuals, there may be a path that evolution takes, to improve its fitness in the huge landscape of all possible genomes. 
There may also be *game interactions*, i.e. concepts from game theory are important. And there can be examples in populations of 'psychopathic' individuals.
